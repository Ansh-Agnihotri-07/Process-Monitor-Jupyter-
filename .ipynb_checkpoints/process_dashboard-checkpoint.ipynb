# Process Monitor Dashboard

Interactive dashboard for monitoring and controlling OS processes using `ipywidgets` and the C++ backend.

```python
import pandas as pd
import time
import os
import sys
from IPython.display import display, clear_output
import ipywidgets as widgets

# Ensure project root is in sys.path
project_root = os.path.abspath(".")
if project_root not in sys.path:
    sys.path.insert(0, project_root)

from gui.notebook_interface import *

print("✅ Imports successful. Ready to monitor processes.")
```

```python
# 2. GLOBAL STATE
selected_pid = None
auto_refresh = True
refresh_interval = 3
```

```python
# 3. PROCESS TABLE FUNCTION
def get_filtered_df():
    # Use existing functions from notebook_interface
    processes = fetch_processes()
    df = processes_to_dataframe(processes)
    df = filter_processes(df)
    df = sort_processes(df)
    
    return df

# 4. PROCESS SELECTOR
def update_dropdown(df):
    if df.empty:
        dropdown.options = [("No processes found", None)]
        return
        
    options = [
        (f"{row['Name']} (PID {row['PID']}) - CPU {row['CPU']}%", row['PID'])
        for _, row in df.iterrows()
    ]
    
    dropdown.options = options
```

```python
# 5. CREATE WIDGETS
dropdown = widgets.Dropdown(description="Process:", layout={'width': 'max-content'})
kill_btn = widgets.Button(description="Kill", button_style='danger', icon='trash')
pause_btn = widgets.Button(description="Pause", icon='pause')
resume_btn = widgets.Button(description="Resume", icon='play')
priority_slider = widgets.IntSlider(min=-10, max=10, value=0, description="Priority")
set_priority_btn = widgets.Button(description="Set Priority", button_style='info')

auto_toggle = widgets.ToggleButton(value=True, description="Auto Control", button_style='warning', icon='robot')
status_label = widgets.Label(value="")
```

```python
# 6. ACTION FUNCTIONS
def on_kill(b):
    pid = dropdown.value
    if pid:
        kill_process(pid)
        status_label.value = f"✅ Killed PID {pid}"
    else:
        status_label.value = "❌ No process selected"

def on_pause(b):
    pid = dropdown.value
    if pid:
        pause_process(pid)
        status_label.value = f"⏸️ Paused PID {pid}"

def on_resume(b):
    pid = dropdown.value
    if pid:
        resume_process(pid)
        status_label.value = f"▶️ Resumed PID {pid}"

def on_priority(b):
    pid = dropdown.value
    if pid:
        val = priority_slider.value
        set_priority(pid, val)
        status_label.value = f"⚡ Priority set to {val} for PID {pid}"

kill_btn.on_click(on_kill)
pause_btn.on_click(on_pause)
resume_btn.on_click(on_resume)
set_priority_btn.on_click(on_priority)
```

```python
# 7. DISPLAY LOOP
def run_dashboard():
    try:
        while True:
            df = get_filtered_df()
            update_dropdown(df)
            
            clear_output(wait=True)
            
            # Layout UI
            ui = widgets.VBox([
                widgets.HTML("<h2>Process Monitor Control Panel</h2>"),
                dropdown,
                widgets.HBox([kill_btn, pause_btn, resume_btn]),
                widgets.HBox([priority_slider, set_priority_btn]),
                auto_toggle,
                status_label
            ])
            
            display(ui)
            
            # Use the styling function from notebook_interface if available, 
            # or just display the DataFrame
            try:
                # Access internal styling if possible, otherwise use standard display
                from gui.notebook_interface import _style_dataframe
                display(_style_dataframe(df))
            except ImportError:
                display(df)
            
            # AUTO CONTROL LOGIC
            if auto_toggle.value:
                for _, row in df.iterrows():
                    # Auto-pause high CPU processes (> 40%)
                    if row["CPU"] > 40.0:
                        pause_process(row["PID"])
            
            time.sleep(refresh_interval)
    except KeyboardInterrupt:
        print("🛑 Dashboard stopped.")
    except Exception as e:
        print(f"❌ Error in dashboard loop: {e}")
```

```python
# 8. RUN
run_dashboard()
```
